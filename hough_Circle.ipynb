```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
def load_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Cannot load image: {image_path}")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img, gray

def detect_circles(gray_image):
    gray_blur = cv2.GaussianBlur(gray_image, (9, 9), 1.5)
    circles = cv2.HoughCircles(
        gray_blur,
        cv2.HOUGH_GRADIENT,
        dp=1.2,
        minDist=60,
        param1=100,
        param2=25,
        minRadius=20,
        maxRadius=40
    )
    return circles

def draw_circles(img, circles):
    """วาดวงกลมและจุดศูนย์กลางลงบนภาพ"""
    result = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    count = 0
    if circles is not None:
        circles = np.uint16(np.around(circles))
        count = len(circles[0])
        for x, y, r in circles[0]:
            cv2.circle(result, (x, y), r, (255, 0, 0), 2)   # วงกลม
            cv2.circle(result, (x, y), 3, (0, 255, 0), -1) # ศูนย์กลาง
    return result, count


def show_results(original, gray, circle_img, count):
    """แสดงผล Original / Grayscale / Circle Detection"""
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
    plt.title("Original")
    plt.axis("off")
    
    plt.subplot(1, 3, 2)
    plt.imshow(gray, cmap="gray")
    plt.title("Grayscale")
    plt.axis("off")

    plt.subplot(1, 3, 3)
    plt.imshow(circle_img)
    plt.title(f"Circle Detection (Count = {count})")
    plt.axis("off")

    plt.show()
```
